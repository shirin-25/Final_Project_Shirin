```python
df_wifi['Date'] = pd.to_datetime(df_wifi['Date heure début'], format="%Y-%m-%d %H:%M:%S", utc=True, exact = False)
```

```python
df_wifi['Date'] = df_wifi['Date'].dt.tz_localize(None)
```

```python
df_wifi = df_wifi[['Code Site', 'Date', 'Etat du site', 'Temps de sessions en minutes',
                            'Donnée entrante (MégaOctet)',
                            'Donnée sortante (MégaOctet)','Paquets entrants',
                            'Paquets sortants', 'Nombre de bornes']]
```

```python
df_wifi['Date'] = df_wifi['Date'].dt.strftime('%Y-%m-%d')
```

```python
df_wifi = df_wifi.sort_values(by="Date")
```

```python
df_wifi
```

```python
df_wifi_per_day = df_wifi.groupby(['Date']).agg({"Code Site": "nunique", 
                                                 'Temps de sessions en minutes' : np.sum,
                                                'Donnée entrante (MégaOctet)' : np.sum,
                                                'Donnée sortante (MégaOctet)' : np.sum,
                                                 'Paquets entrants' : np.sum,
                                                 'Paquets sortants' : np.sum,
                                                 'Nombre de bornes' : np.sum,
                                                })
```

```python
df_wifi_per_day_for_display = df_wifi_per_day.reset_index()
```

```python
df_wifi_per_day_for_display.plot.line(x='Date', y='Code Site')
```

```python
# the target is the number of sites used per day
```

```python
X = df_wifi_per_day.drop(['Code Site'], axis = 1)
y = df_wifi_per_day['Code Site']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.15)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()

lr.fit(X_train, y_train)

y_pred_train = lr.predict(X_train)

y_pred_test = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

mae_train = mean_absolute_error(y_train , y_pred_train)

mae_test = mean_absolute_error(y_test, y_pred_test)

print(mae_train)

print(mae_test)

mean_code = df_wifi_per_day['Code Site'].mean()

print("\nRelative error", mae_test/mean_code)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score_test = r2_score(y_test, y_pred_test)
```

```python
print(r2_score_test)
```

```python
model = LinearRegression().fit(X_train, y_train)

# Get predictions for the test set.
y_predict = model.predict(X_test)

mean_squared_error = mean_squared_error(y_test, y_predict)
print('Mean Squared Error: %.2f' % mean_squared_error)

r2_score = r2_score(y_test, y_predict)
print('R2 Score: %.2f' % r2_score)

# As a custom metric, we're curious to check how many of the
# predictions were within 20% of the true prediction.
percent_diff = np.abs(y_predict - y_test) / y_test
result = (percent_diff < 0.2).sum() / len(y_test)
print('Percent within 20%% of target value: %.2f' % result)
```
